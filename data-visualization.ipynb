```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Load and inspect the dataset.

```python
df=pd.read_csv("sales_data.csv")
```

2. Visualize the sales distribution across product categories using a bar plot

```python
df.head()
```

```python
plt.bar("Product","Category",data=df,color="red",width=0.2)
plt.xlabel("Product")
plt.ylabel("Category")
plt.title("Product vs Category")
plt.show()
```

3. visualize the relationship between sales and profit using a scatter plot

```python
plt.scatter("Sales","Profit",data=df,s=150)
plt.xlabel("Sales")
plt.ylabel("Scatter")
plt.title("sales vs profit")
plt.show()
```

```python
df.head()
```

4. create a histogram to display the distribution of quantity sold

```python
plt.hist("Quantity",10,histtype='bar')
plt.title("Quantity sold")
plt.show()
```

5. generate a box plot to compare discounts given in each region

```python
sns.boxplot(x="Region",y="Discount",data=df)
plt.title("discount distribuion by region")
plt.xlabel("region")
plt.ylabel("discount")
plt.show()
```

6. create a heatmap showing the correlation between different numerical variables.

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt=".2f")
plt.title("correlation heatmap")
plt.show()
```
